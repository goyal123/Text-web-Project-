```python
import glob
import string
import pandas as pd
```

```python
files = glob.glob(r'E:\Project\text and web\articles1\*.txt')
data = {}
for name in files:
    filename = name.split('\\')[-1].split('.')[0]
    with open(name,encoding="utf8") as f:
        value = f.read().count('http')
        print(filename, value)
        data[filename]= value
print(data)  
```

```python
df = pd.read_csv('new_records.csv', encoding='iso-8859-1')
```

```python
counts = [] 
for title in df['Title']:
    title = [ i for i in title if i not in string.punctuation+' ']
    title = ''.join(title)
    counts.append(data[title])

df['Link Count'] = counts
```
