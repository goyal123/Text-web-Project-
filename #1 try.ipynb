```python
def func(f):
    mlist = []
    file = f.read()
    word = nltk.word_tokenize(file)
    NoWord = [',','(',')',':',';','.','%','\x96','{','}','[',']','!','?',"''","``"]
    word2 = [i for i in word if i not in NoWord]
    clean_tokens = word2[:]
    for token in word2:
        if token in stopwords.words('english'):
            clean_tokens.remove(token)
    return len(set(clean_tokens)) 
```
